In this project, we aim to train a perceptron classification model on our dataset. After training, the perceptron will provide its coefficients and intercept, which are essential for defining the decision boundary separating different classes in the feature space.

```python
from sklearn.datasets import make_classification
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
xip,y=make_classification(n_samples=100,n_features=2,n_classes=2,n_informative=1,n_redundant=0, random_state=41,n_clusters_per_class=1,hypercube=False,class_sep=10)
```

```python
sns.scatterplot(x=xip[:,0],y=xip[:,1],hue=y)
```

```python
def step(a):
    if a>0:
        return 1
    else:
        return 0
```

```python
def perceptron(x,y,epochs):
    x=np.insert(xip,0,1,axis=1)
    w=np.random.rand(1,x.shape[1])
    neta=0.1
    for i in range(epochs):
        a=np.random.randint(x.shape[0])
        xi=x[a,:]
        w=w+neta*xi*(y[a]-step(np.dot(w,xi)))
    return w
```

```python
w=perceptron(x,y,1000)
```

```python
x1=np.linspace(-2.5,2.5,100)
y1=-(w[0][1]/w[0][2])*x1-(w[0][0]/w[0][2])
sns.scatterplot(x=xip[:,0],y=xip[:,1],hue=y)
plt.plot(x1,y1)
plt.xlim(-3,3)
plt.ylim(-5,5)
```
