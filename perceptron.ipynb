# Perceptron model

Import all the required libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
```

The dataset comprises information on students' CGPA (Cumulative Grade Point Average) and resume scores, which are utilized for predicting whether they will secure placements or not.

```python
df=pd.read_csv("datasets\placement.csv")
```

```python
df.shape
```

```python
df.head(5)
```

The relationship depicted below suggests that a linear model can be effectively applied to the data.

```python
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

Split the train and test data

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
p=Perceptron()
```

```python
p.fit(x,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
plot_decision_regions(x.values,y.values,clf=p)
```

Due to the limited amount of training data available, the performance of the model did not meet expectations.
